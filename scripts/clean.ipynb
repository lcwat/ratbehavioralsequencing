```python
import pandas as pd
from glob import glob
```

```python
file_list = glob('..\\data\\aged_rats\\sample_phase\\*')
```

```python
# loop through file list, clean, then write to file
for file in file_list:
    df = pd.read_csv(file, header=None, index_col=0)

    # id each unique frame chunk for behavior grouping
    behavior_frames = []

    # loop through frames and group together with id nums
    for i in range(len(df)):

        if i == 0:
            counter = 1
        elif df.iloc[i, 0] != df.iloc[i-1, 0]:
            counter += 1

        behavior_frames.append(counter)

    df['b'] = behavior_frames

    # parse file name for id
    id = file.split('\\')[4].split('_')[0]
    group = file.split('\\')[2].split('_')[0]
    phase = file.split('\\')[3].split('_')[0]

    # add extra info
    df['id'] = [id for i in range(len(df))]
    df['group'] = [group for i in range(len(df))]
    df['phase'] = [phase for i in range(len(df))]

    df.columns = ['behavior', 'behavior_id', 'animal_id', 'group_id', 'phase']

    df.to_csv(
        '../data/cleaned/'+group+'_'+phase+'_'+id+'_behavior_seq.csv',
        index=False
    )
```
