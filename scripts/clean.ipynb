```python
import pandas as pd
from glob import glob
from tqdm import tqdm
```

```python
# get files
file_list = glob('..\\data\\test_phase_novel_and_familiar\\young\\*')

# view
print(file_list[0:4])
```

```python
# loop through file list, clean, then write to file
for file in tqdm(file_list):
    df = pd.read_csv(file)

    # rename
    df.columns = ['f', 'behavior']

    # id each unique frame chunk for behavior grouping
    behavior_frames = []

    # loop through frames and group together with id nums
    for i in range(len(df)):

        if i == 0:
            counter = 1
        elif df.iloc[i, 1] != df.iloc[i-1, 1]:
            counter += 1

        behavior_frames.append(counter)

    df['behavior_id'] = behavior_frames

    # parse file name for id, uncomment if test novel/familiar data
    id = file.split('\\')[4].split('_')[0]
    group = file.split('\\')[3]

    # uncomment if simple sample
    # id = file.split('\\')[5].split('_')[0]
    # group = file.split('\\')[3]
    # phase = file.split('\\')[4]

    # add extra info
    df['animal_id'] = [id for i in range(len(df))]
    df['group_id'] = [group for i in range(len(df))]
    # df['phase'] = [phase for i in range(len(df))]  # uncomment if simple_sample
    df['phase'] = ['test' for i in range(len(df))]

    df = df.drop(columns='f')

    df.to_csv(
        '../data/cleaned/test_novel_fam/'+group+'_'+'test'+'_'+id+'_behavior_seq.csv',
        index=False
    )

    # df.to_csv(
    #     '../data/cleaned/simple_sample_test/'+group +
    #     '_'+phase+'_'+id+'_behavior_seq.csv',
    #     index=False
    # )
```
